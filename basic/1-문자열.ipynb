```python
#### 비트 연산자

a = 7
b = 3

result = a>3 and b<5 # 하나만 거짓이여도 거짓

print(result)

result = a>3 & b<5 # 비트 연산자
print(result)

result = a & b    # 111
print(result)     # 011
                  # 011
```
