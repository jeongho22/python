### (2) while

    while 조건식 : 
        문장
        ...
        끝나는 시점을 만들어줘야함 아니면 무한반복임
        

```python
print("**********")
print("*"*10)
print("s"*15)

cnt = 0 
while cnt < 10:
    print("e", end="") 
    cnt = cnt + 1 # 꼭 해줘야함 안그러면 무한반복.
    

print("\n--------------------------------------\n")    
    
cnt = 0
while cnt < 5:
    print("유정" ,end="")
    cnt = cnt+1
```

```python
###### 1부터 10까지의 합계

cnt = 1
sum = 0

while cnt < 10:
    sum = sum+ cnt
    
    cnt = cnt+1
    
    print("결과: ",sum)
```

```python
###### 2의 10승 : 1024

cnt = 0 
sum = 1 #초기값


while cnt < 10:
    sum *=10
    cnt= cnt+1
    
    print("결과:" ,sum)
```

### (3) for

        문장

```python
data = [10,20,"홍길동",3.14,False]

cnt = 0 
while cnt< 5: 
    print(data[cnt])
    cnt += 1

print("-------------------")

for v in data :
      print(v)
        
print("-------------------")

for v in ["유정","정호","섹스",1,1,1,1,"2"]:
       print("3", end="")
        
print()   
    
print("---------------------------")

for v in range(826) : 
    print("3", end="")
```

### (5)반복문의 중첩

```python
"""
**********
**********
**********
"""

for v1 in range(5): # 행의 개념
    for v2 in range(5): # 열의 개념
        print("A",end="")
    print()
```

```python
"""
*
**
***
****
*****
******
"""

for v1 in range(6): # 행의 개념
    for v2 in range(v1+1): # 열의 개념
        print("*",end="")
    print()
 

    ##6 줄이긴한데 0이 생략 되어있어서 그런거임. 근데 거기에 +1 해주닌깐 1,2,3,4,5,6
    
    


    
for v1 in range(6):
    for v2 in range(v1):
        print("*", end = "")
    print()
    
    
    
    
    # 0 이 생략되었음 왜 근데 5개냐? v1+1 을 안해줘서그럼.
```

```python
"""
     *
    ***
   *****
  *******
 *********
"""

for v1 in range(6):
    for v2 in range(v1*2-1):
        print("*", end = "")
    print()
    
```

(6) List Comprehension 

[o,o,o,o,o,o,o,o]
[반복문.........]


```python
a=[]

for i in range(10):
        a.append(i)
    
print(a)


################################

a = [i for i in range(10)] # 어펜드 생략한 i 를 넣어주면 차곡차곡 쌓인다. !
print(a)

print("\n--------------------------\n")

a = []

for i in range(10):
    if i % 2!= 0:
        a.append(i)
        
print(a)

##############################

a = [ i for i in range(10) if i % 2 == 0 ]
print(a)
```

```python
c1 = ["A","B","C"]

c2 = ["D","E","A"]

# ["AD","AE","AA","BD","BE","BA","CD","CE","CA"]

result = []
for row in c1:
    for col in c2:
        result.append(row + col)
        
        print(result)


# for row in c1 한번돌때 for col in c2 세번돈다

result = [row+col for row in c1 for col in c2]
print(result)


# ["AD","AE","AA","BD","BE","BA","CD","CE","CA"]


result = []
for row in c1:
    for col in c2:
        if row !=col:
            result.append(row + col)   
    
print(result)

result = [row+col for row in c1 for col in c2 if not row == col]
print(result)


print("\n-----------------------------\n")

# [["AD","AE","AA"],[]"BD","BE","BA"],[]"CD","CE","CA"]]
```
