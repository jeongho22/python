# 11. 배열 연결(병합) : stack(),hstack(),dstack(),title()

```python
#### hstack(), vstack()

arr1 = np.ones((2,3))
arr2 = np.ones((2,2))

print(arr1)
print(arr2)

arr3 = np.hstack([arr1,arr2])

print(arr3)

## 행으로 병합

# arr3 = np.vstack([arr1,arr2]) 행렬 갯수를 보고써야한다.
```

```python
##### dstack() # 맨끝에 축을하나 더 세워둔다. 차원을 하나 더 증가시켜준다. 하나씩 밀린다

arr1 = np.ones((3,4)) 
arr2 = np.ones((3,4))
arr3 = np.ones((3,4))



print(arr1)
print(arr2)
print(arr3)
print(arr3.shape)


print("----------------------")

# 3,4,3
arr4 = np.dstack([arr1,arr2,arr3])
print(arr4)
print(arr4.shape)
```

```python
#### stack()
# axis = 새로운 축을 세울 위치
            # 0 1 2
arr1 =np.ones((3,4))
arr2 =np.ones((3,4))

arr3 =np.stack([arr1,arr2])
print(arr3)
print(arr3.shape)

print("----------------------")

arr3 = np.stack([arr1,arr2],axis=0)

print(arr3)
print(arr3.shape)

print("----------------------")

arr3 = np.stack([arr1,arr2],axis=1)

print(arr3)
print(arr3.shape)

print("----------------------")

arr3 = np.stack([arr1,arr2],axis=2)

print(arr3)
print(arr3.shape)
```

```python
#### tile()

arr = np.array([[0,1,2],[3,4,5]])
print(np.tile(arr,2))

print("----------------------------")

print(np.tile(arr,(3,2)))
```

# 18.정수형 데이터 카운팅: unique(),bincount()

```python
#### unique
arr = np.array(["a","b","b","c","a"])
print(len(arr))
print(np.unique(arr))
print(len(np.unique(arr)))


print("-------------------------------")

idx, cnt = np.unique(arr, return_counts=True) # 중복된 갯수랑 문자를 알려줌
print(idx,cnt)

```

```python
#### bincount : 발생하지 않은 사건에 대해서도 카운팅

arr = np.array([1,1,2,2,2,3])

print(np.unique(arr, return_counts=True))
print(np.bincount(arr,minlength=6)) 
# 0은 안나왔고 1은 2번 나왔고 2는 3번 나왔고 3은 1번 나왔다 4,5는 나오지않았다
```

# (5)apply()

+ 행이나 열 단위로 더 복잡한 처리를 하고자 할 때 사용
+ 인수로 행 또는 열을 받는 함수를 apply()의 인자로 넣으면 각 열 (또는 행)을 반복하여 함수의 적용

```python
df = pd.DataFrame({
         "A" :[1,3,4,3,4],
         "B" :[2,3,1,2,3],
         "C" :[1,5,2,4,4]
})
df
```

```python
##### 각 열(행)별로 최대값에 최소값을 뺀 값을 구하고자 한다.

def diff(x):
    return x.max() - x.min()

print(diff(df["A"]))
print(diff(df["B"]))
print(diff(df["C"]))

print("--------------------------------")

print(df.apply(diff, axis=0)) # 위에 3개를 한번에 해주는 시스템
print(df.apply(diff, axis=1)) # 행의 합계
```

```python
##### 람다 함수 (이름이 없는함수 라는뜻 1번만 사용하기위한 함수)

df.apply(lambda x:x.max()-x.min(),axis=0)
```

```python
##### 각 열에 대해 어떤 값이 얼마나 사용되었는지 조회

print(df["A"].value_counts())
print(df["B"].value_counts())
print(df["C"].value_counts())



df.apply(lambda x:x.value_counts(), axis=0)

#df.apply(pd.value_counts, axis=0)
```

```python
########### apply(),agg(),transform()
#agg 동시에

import seaborn as sns
iris = sns.load_dataset("iris")
iris.head()
```

```python
# 각 품종별로 가장 큰 값과 가장 작은값의 비율을 조회

def peek_to_peek_ratio(x) : 
    return x.max()/x.min()

##################################################

g = iris.groupby("species")

print(g.apply(peek_to_peek_ratio))

print("------------------------------------------")

print(g.agg(peek_to_peek_ratio))

print("------------------------------------------")

print(g.transform(peek_to_peek_ratio))
```

```python


##### 연습 문제

df = pd.DataFrame({
    'city': ['부산', '부산', '부산', '부산', '서울', '서울', '서울'],
    'fruits': ['apple', 'orange', 'banana', 'banana', 'apple', 'apple', 'banana'],
    'price': [100, 200, 250, 300, 150, 200, 400],
    'quantity': [1, 2, 3, 4, 5, 6, 7]
})

print(df)

print("===========================================================")

# 1) 도시별로 과일의 가격 평균과 수량 평균을 구하시오.

df.groupby(["city"]).mean()[["price","quantity"]]


# 2) 도시별, 과일별 가격 평균과 수량 평균을 구하시오.

df.groupby(["city","fruits"]).mean()[["price","quantity"]]

# 3) 위의 문제에서 도시별, 과일별을 인덱스로 하고 싶지 않은 경우

df.groupby(["city","fruits"],level= 0).mean()[["price","quantity"]]

# 4) 도시별로 가격의 평균과 수량의 합계를 동시에 구하시오.

df.groupby(["city"]).agg([np.sum,np.mean])

# 5) 과일별로 도시의 가격 평균과 수량 평균을 구하시오.

df.groupby(["fruits","city"]).agg([np.mean])
```

```python
##### shift 연산

np.random.seed(0)

ts = pd.Series(np.random.randn(4),index=pd.date_range("2021-1-3",periods=4,freq="M"))
print(ts)

print("----------------------------------------------------------------------------")

print(ts.shift(1))

print("----------------------------------------------------------------------------")

print(ts.shift(-1))

print("----------------------------------------------------------------------------")

print(ts.shift(1,freq="M")) # 인덱스 이동

print("----------------------------------------------------------------------------")

print(ts.shift(1,freq="W")) # 첫쨋주 일요일

```

```python
#### resampling : 시간 간격을 다시 재조정
# 시간 구간이 작아지면 데이터양이 증가 : 업 생플링
# 시간 구간이 작아지면 데이터양이 감소 : 다운 생플링

ts = pd.Series(np.random.randn(100), index=pd.date_range("2021-1-1", periods=100,freq="D"))
ts.tail(10)

# 다운 샘플링 : 데이터가 그룹으로 묶이기 때문에 집계 연산을 해서 대표값을 구해야한다.
print(ts.resample("W").mean()) # 일요일
print("-------------------------------------------------------------------------------")
print(ts.resample("M").mean()) # 각 달의 마지막달
print("-------------------------------------------------------------------------------")
ts = pd.Series(np.random.randn(60),index=pd.date_range("2012-1-1",periods=60, freq="T"))
ts
print(ts.resample("20T").sum()) # 20초마다

```

```python
####### up-sampling

# forward filling : 앞에서 나온 데이터를 그대로 사용
# backward filling : 뒤에서 나올 데이터를 앞에서 미리 사용

print(ts.head())
print("---------------------------------")

print(ts.resample("30S").ffill().head(10))  #  -0.480242,-0.480242
print("---------------------------------")
print(ts.resample("30S").bfill().head(10))
```
