```python
jot_sung_love = pd.DataFrame({    
    'x': [[30,1],[14,2],[14,3],[14,4],[14,5],[14,6]],
    'y': [[100],[0],[0],[0],[0],[0]]})

jot_sung_love
```

```python
# 썸 탄 기간 : 독립변수
x = [[30,1],[14,2],[14,3],[14,4],[14,5],[14,6]]   # [썸 탄 기간(일 수),횟수(썸포함)] 일반적으로 썸 기간 14일로 잡음. 
                                                  # 단, 좆성이의 첫번째 연애는 썸탄 기간을 30일로 잡아줌.
    
    
# 연애 성공 : 종속변수 
y = [[100],[0],[0],[0],[0],[0]]                   # 사귀면 연애성공 [100] 으로 채워줌. 썸타고 안사귄건 [0] 으로 채워줌



# 학습시키기
from sklearn.linear_model import LinearRegression
model = LinearRegression()                        # 선형회귀분석 객체 생성하기



                            
model.fit(x,y)                                    # 객체 데이터를 기반으로 선형회귀분석을 통한 연애 학습 훈련 시키기


# 예측하기
result = model.predict([[15,7]])                  # 썸탄 기간이 15일때 -> 다음 일곱번째 연애확률 예측 해보기
print("예상확률:", result)
```
