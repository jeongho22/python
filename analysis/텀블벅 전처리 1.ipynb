```python
list_of_single_column = list(df_INNER_JOIN['상품명'])              ##### df1 : 성사된 / df2: 진행중
list_of_single_column  #겹치는 상품 리스트 타입으로 변환

df_final = df1[~df1['상품명'].isin(list_of_single_column)] # 성사된 프로젝트만 추출 완료
df_final
```

## 목표금액 "배송 완료"  결측치 대체

```python
df_final.to_csv("data/df_final.csv", encoding = "utf-8")
```

```python
df1 = pd.read_csv("data/df_final.csv", index_col = 0, thousands = ',')
```

```python
df1.info()
```

```python
df1["목표금액"] = df1["목표금액"].str.replace(',', '')
```

```python
df1
```

```python
new_wish = df1["현재 펀딩 금액"] / df1["펀딩 퍼센트"]

for i in range(len(df1["목표금액"].values)):         # i는 순번
    if df1["목표금액"].values[i] == "배송 완료":
        df1["목표금액"].values[i] = new_wish.values[i].astype('int')
    else: 
        pass

df1["목표금액"] = df1["목표금액"].values
```
