# (1) 퍼셉트론
   + 다수의 신호를 입력으로 받아 하나의 신호로 출력

# 1) AND gate

```python
def AND(x1 , x2):
    w1,w2 = 0.5,0.5
    b = 0.7
    
    tmp = w1*x1+w2*x2
    
    if tmp<= b:
        return 0
    else: 
        return 1
```

```python
inputs = [(0,0),(1,0),(0,1),(1,1)]

for x1,x2 in inputs:
    y = AND(x1,x2)
    print("{},{} ->{}". format(x1,x2,y))
#         0       0     0      1
```

# 2) OR gate

```python
def OR(x1 , x2):
    w1,w2 = 0.5,0.5
    b = 0.2
    
    tmp = w1*x1+w2*x2
    
    if tmp<= b:
        return 0
    else: 
        return 1
```

```python
inputs = [(0,0),(1,0),(0,1),(1,1)]

for x1,x2 in inputs:
    y = OR(x1,x2)
    print("{},{} ->{}". format(x1,x2,y))
#         0       0     0      1
```

### 3) XOR gate

```python
def NAND(x1,x2):
    w1,w2 = -0.5,-0.5
    b = 0.7
    
    tmp = (w1*x1+w2*x2)+b
    
    if tmp <= 0:
        return 0
    else: 
        return 1
    
```

```python
inputs = [(0,0),(1,0),(0,1),(1,1)]

for x1,x2 in inputs:
    y = NAND(x1,x2)
    print("{},{} ->{}". format(x1,x2,y))
#         0       0     0      1
```

```python
def NAND(x1,x2):
    x = np.array([x1,x2])
    w = np.array([-0.5,-0.5])
    b = 0.7
    
    tmp = np.sum(w*x)+b
    
    if tmp <= 0:
        return 0
    else: 
        return 1
    
```

```python
inputs = [(0,0),(1,0),(0,1),(1,1)]

for x1,x2 in inputs:
    y = NAND(x1,x2)
    print("{},{} ->{}". format(x1,x2,y))
#         0       0     0      1
```

```python
def XOR(x1,x2):
    s1 = NAND(x1,x2)
    s2 = OR(x1,x2)
    y = AND(s1,s2)
    
    return y

```

```python
inputs = [(0,0),(1,0),(0,1),(1,1)]

for x1,x2 in inputs:
    y = XOR(x1,x2)
    print("{},{} ->{}". format(x1,x2,y))
#         0       0     0      1
```
