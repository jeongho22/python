2. 머신러닝 종류 
=====================
(1) 지도 학습(교사 학습) : Supervisor Learning
        1) 회귀 (Regression)
            -선형회귀
            -릿지회귀, 라쏘회귀, 엘라스틱넷회귀
        
        2) 분류(Classification)
            - Logistic Regression : 독립변수와 종속변수의 선형 관계성을 기반으로 하는 알고리즘
            - KNN : 근접 거리를 기준으로 하는 근접 알고리즘
            - Decision Tree : 데이터의 균일도에 따른 규칙 기반 알고리즘
                    - Random Forest
                    - GBM : XGBoost, LightGBM
            - SVM : 개별 클래스간의 최대 분류 마진을 효과적으로 찾아주는 알고리즘
            - Naive Bayes : 베이즈 통계와 생성 모델을 기반으로 하는 알고리즘
            - Neural Network :심층 연결 기반의 신경망
        
        


(2) 비지도 학습(비교사 학습) : Unsupervisor Learning
    1) 군집(Clustering)
        - K Means
        - DBSCAN
        - 이상치 탐지,특이치 탐지

    2) 시각화, 차원축소
    
    
    3) 연관 규칙 학습(Association Rule Learning)
    
    
    4) 피처 추출(Feature Extraction)

(3) 강화 학습 (로봇청소기같은..) : Reinforce Learning
        -완전한 답을 제공하지 않는다.
        -보상 시스템 적용 
        -대표적인 예 : 알파고

### (7) 응용 예제 2 : 영화 분류 프로그램

```python
import operator

def createDataSet():
    group = np.array([[1.0,2.0],[1.0,4.0],[4.0,1.0],[4.0,2.0]])
    label = ["Action","Action","Romance","Romance"]
    return group, label


# target 목표

def calcDistance(X_new,data):
    # dataSize = len(data)
    dataSize = data.shape[0]
    # 새로 입력된 데이터와 기존 데이터의 차이
    diffMat = np.tile(X_new,(dataSize,1))- data
    
    # L2 norm
    sqDiffMat = diffMat**2
    distance = (sqDiffMat.sum(axis=1)) ** 0.5 # 행 =1 =0 열
    return distance.argsort()
    
    

def findMajorityClass(sortedDistance,target,k):
    classCount ={}
    
    for i in range(k):
        voteLabel = target[sortedDistance[i]]
        classCount[voteLabel] = classCount.get(voteLabel,0) + 1
        
    return sorted(classCount.items(), key=operator.itemgetter(1),reverse=True)
    

def classify(X_new, data,target, k):
    #거리 측정
    sortedDistance = calcDistance(X_new,data)
    
    # 이웃의 갯수로 묶어서 해당 클래스로 분류
    result = findMajorityClass(sortedDistance,target,k)
    return result
```

```python
X, y = createDataSet()
# print(X, y)

result = classify([2.0, 3.0], X, y, 3)
print("결과는 {} 영화이다.".format(result[0][0]))
```

# (2) Cross Validation

+ K fold 방식

```python
from sklearn.model_selection import KFold
```

```python
kfold=KFold(n_splits=5) # 나누는거 5등분
list(kfold.split(iris.data))

# 하나는 테스트용 나머지는 훈련용
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

clf = KNeighborsClassifier(n_neighbors=3)

cv_accuracy = []
cnt = 0 

for train_index, test_index in kfold.split(iris.data):
    #print("훈련용:", train_index)
    #print("테스트용:", test_index)
    X_train, X_test = iris.data[train_index],iris.data[test_index]
    y_train, y_test = iris.target[train_index],iris.target[test_index]
    #print(X_train.shape, X_test.shape,y_train.shape,y_test.shape)
    
    
    # 학습 및 예측 테스트
    clf.fit(X_train,y_train)
    pred = clf.predict(X_test)
    cnt = cnt + 1
    
    # 정확도 측정
    accuracy = np.round(accuracy_score(y_test,pred),4)
    print("{}회 검증 - 정확도:{}, 학습 데이터 크기{},검증 데이터크기{}".format(cnt,
                                                           accuracy,X_train.shape[0],
                                                          X_test.shape[0]))
    
    
    cv_accuracy.append(accuracy)
    
print("\n\n 평균 검증 정확도:",np.mean(cv_accuracy))
```

### (3) GridSearchCV

+ 교차 검증과 최적의 하이퍼 파라미터 튜닝을 한번에 처리

+ 파라미터
    - estimator : classifier, regressor, pipeline
    - param_grid : key + 리스트값을 가지는 dict
    - scoring : 평가지표
    - cv : 교차 검증을 윟 분할되는 학습 / 테스트 셋의 갯수를 지정
    - refit : 디폴트가 True, 최적의 하이퍼 파라미터를 찾은 뒤 재학습 여부

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.datasets import load_iris
iris = load_iris()

X_train, X_test, y_train, y_test = train_test_split(iris.data,iris.target,
                                                    random_state=121,test_size=.2)

knn = KNeighborsClassifier()

```

```python
params = {"n_neighbors": [n for n in range(1,15) if n%2 != 0]}

grid_knn = GridSearchCV(knn, param_grid=params, cv=3)
grid_knn.fit(X_train, y_train)
```

```python
scores_df = pd.DataFrame(grid_knn.cv_results_)
scores_df[["params","mean_test_score","rank_test_score","split0_test_score",
              "split1_test_score","split2_test_score"]]
```

```python
from sklearn.metrics import accuracy_score

print("최적의 피라미터 : ", grid_knn.best_params_)
print("최적 정확도 : ", grid_knn.best_score_)

pred= grid_knn.predict(X_test)
print("테스트 정확도:" , accuracy_score(y_test,pred))
```
