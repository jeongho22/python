# 4. 데이터 전처리 

+ 데이터를 X,y로 분리
+ Scaling

### (1) X와 y로 분리


```python
X = data.drop(columns=["Pass/Fail"], axis=1) # pass/fail 컬럼삭제
y = data["Pass/Fail"] # 따로 y에 넣어놓음

X
y = y.to_numpy()
y
```

```python
data_test = pd.read_csv("data1/uci-secom-test.csv")
data_test


X_test = data_test.drop(columns=["Pass/Fail"], axis=1)
y_test = data_test["Pass/Fail"].to_numpy()
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

X_train = sc.fit_transform(X)
X_test = sc.transform(X_test)
y_train = y
```

```python
print(X_train.mean())
print(X_train.var())
```

# 5.머신러닝 모델 학습

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=5000)

model.fit(X_train,y_train)

print(model.score(X_train,y_train)) # 훈련
print(model.score(X_test,y_test)) # 테스트

#과적합으로 보임
```

```python
?LogisticRegression
```

```python
### 변수 중요도 확인

abs_coef = np.abs(model.coef_).ravel()
abs_coef
```

```python
### 변수의 중요도 확인 59번데이터!


lr_import_x = [str(i[0]) for i in sorted(enumerate(abs_coef), key=lambda x:x[1], reverse=True)]
lr_import_x

plt.bar(lr_import_x[:30],sorted(abs_coef, reverse=True)[:30])
plt.rcParams["figure.figsize"]= (15,10)
plt.xlabel("Features")
plt.ylabel("Weight absolute values")
```

```python
### 만약 30번째 센서의 가중치값을 알고 싶을 때
print(sorted(abs_coef, reverse=True)[30]) #30번째 중요도
 
print(lr_import_x[30]) # 30번째중요도를 가진건 249번이다!
```

### (2) 다양한 분류 모델 학습

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
import xgboost as xgb
from xgboost.sklearn import XGBClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
models= []
models.append(("KNN", KNeighborsClassifier()))
models.append(("NB", GaussianNB()))
models.append(("CART", DecisionTreeClassifier()))
models.append(("RF" , RandomForestClassifier()))
models.append(("LDA" , LinearDiscriminantAnalysis()))
models.append(("SVM" , SVC(gamma="auto")))
models.append(("XGB" , XGBClassifier()))
```

```python
for name, model in models:
    model.fit(X_train, y_train)
    
    msg = "{} - train.score : {},test score : {}".format(name, model.score(X_train,y_train),
                                                        model.score(X_test, y_test))
    
    print(msg)
```
